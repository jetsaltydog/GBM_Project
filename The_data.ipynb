```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("train.csv")
ID = data['id']

#Drop unnecessary columns
data = data.iloc[:,1:]
```

```python
data['log_loss'] = np.log(data['loss'])
data
```

```python
# Skewness of the distribution. Values close to 0 show less skew

print(data.skew())

# loss shows the highest skew, log_loss shows the lowest skew
```

```python
sns.violinplot(data=data,y="log_loss").set_title("Violin plot of log(loss)", size = 20)
```

```python
def formatting_loss(ax):
    ax.set_xlabel('ID',fontsize = 14 )
    ax.set_ylabel('loss',fontsize = 14)
    ax.tick_params(labelsize=14)
    ax.set_title('Plot of loss',fontsize = 20)
    return ax
```

```python
def formatting_logloss(ax):
    ax.set_xlabel('ID',fontsize = 14 )
    ax.set_ylabel('loss',fontsize = 14)
    ax.tick_params(labelsize=14)
    ax.set_title('Plot of log(loss)',fontsize = 20)
    return ax
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(ID, data['loss'])#, color = 'orange')
ax = formatting_loss(ax)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(ID, data['log_loss'])#, color = 'orange')
ax = formatting_logloss(ax)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))
ax[0].plot(ID, data['loss'])
ax[1].plot(ID, data['log_loss'])
```

```python
size = 15
cols=data.columns


# Correlation tells relation between two attributes.
# Correlation requires continous data. Hence, ignore categorical data

# Calculates pearson co-efficient for all combinations
data_corr = data.corr()

# Set the threshold to select only highly correlated attributes
threshold = 0.5

# List of pairs along with correlation above threshold
corr_list = []

#Search for the highly correlated pairs
for i in range(0,size): #for 'size' features
    for j in range(i+1,size): #avoid repetition
        if (data_corr.iloc[i,j] >= threshold and data_corr.iloc[i,j] < 1) or (data_corr.iloc[i,j] < 0 and data_corr.iloc[i,j] <= -threshold):
            corr_list.append([data_corr.iloc[i,j],i,j]) #store correlation and columns index

#Sort to show higher ones first            
s_corr_list = sorted(corr_list,key=lambda x: -abs(x[0]))


#Print correlations and column names
for v,i,j in s_corr_list:
    print ("%s and %s = %.2f" % (cols[i],cols[j],v))

# Strong correlation is observed between the following pairs
# This represents an opportunity to reduce the feature set through transformations such as PCA
```
